```python
import os
import requests
import pandas as pd
import numpy as np
from PIL import Image
from io import BytesIO
from ultralytics import YOLO
from IPython.display import display
import googlemaps
import polyline
from kaggle_secrets import UserSecretsClient

#  Google Maps API key (store securely in environment variable)
user_secrets = UserSecretsClient()
GMAPS_API_KEY = user_secrets.get_secret("GMAPS_API_KEY")
gmaps = googlemaps.Client(key=GMAPS_API_KEY)


#  Get user input
src = input("Enter source location (e.g., 'MG Road, Bangalore'): ")
dst = input("Enter destination location (e.g., 'Koramangala, Bangalore'): ")

#  Geocode address to (lat, lng)
def geocode(addr):
    result = gmaps.geocode(addr)
    if not result:
        raise ValueError(f"Could not geocode: {addr}")
    loc = result[0]['geometry']['location']
    return (loc['lat'], loc['lng'])

#  Get route polyline from Google Maps Directions API
def get_route_polyline(src, dst):
    directions = gmaps.directions(src, dst, mode='driving', alternatives=False)
    if not directions:
        raise ValueError("Could not fetch directions.")
    return directions[0]['overview_polyline']['points']

#  Get Street View image from a lat/lng and heading
def get_streetview_image(lat, lng, heading=0, pitch=0, fov=90):
    url = (
        f"https://maps.googleapis.com/maps/api/streetview?"
        f"size=640x640&location={lat},{lng}&fov={fov}&heading={heading}"
        f"&pitch={pitch}&key={GMAPS_API_KEY}"
    )
    response = requests.get(url)
    if response.status_code == 200:
        return Image.open(BytesIO(response.content))
    return None

#  Detect street lights in an image using YOLO
def detect_lamps(image):
    results = model(image)
    return len(results[0].boxes)  # Number of detections

#  Process the route and compute lighting score
def process_route(src_addr, dst_addr, interval=300):
    src = geocode(src_addr)
    dst = geocode(dst_addr)
    encoded = get_route_polyline(src, dst)
    coords = polyline.decode(encoded)
    
    sampled_points = coords[::max(1, len(coords) // 8)]  # Sample ~20 points
    total_lamps = 0
    total_images = 0

    print("\n Analyzing street lights at route points...\n")

    for i, (lat, lng) in enumerate(sampled_points):
        lamp_count = 0
        for heading in [0, 90, 180, 270]:
            img = get_streetview_image(lat, lng, heading=heading)
            if img:
                lamps = detect_lamps(img)
                lamp_count += lamps
                total_images += 1
        total_lamps += lamp_count
        print(f"Point {i+1}: Lat {lat:.5f}, Lng {lng:.5f}  {lamp_count} lamps")

    if total_images == 0:
        print("\n No street view images available.")
        return {"lamp_count": 0, "lighting_score": 0.0}

    lighting_score = total_lamps / total_images
    print(f"\n Total Detected Lamp Posts: {total_lamps}")
    print(f" Total Images Analyzed   : {total_images}")
    print(f" Lighting Score (avg lamps/image): {lighting_score:.2f}")

    return {
        "lamp_count": total_lamps,
        "images_analyzed": total_images,
        "lighting_score": lighting_score
    }

#  Run
result = process_route(src, dst)
```
